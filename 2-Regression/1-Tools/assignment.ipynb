```python
# lib imports
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, linear_model, model_selection

# load the linnerud dataset
data = datasets.load_linnerud()

# print what it is inside the dataset
print(data.keys())
```

```python
# name of exercise and physiological variables
print("Feature names (Exercise variables):", data.feature_names)
print("Target names (Physiological variables):", data.target_names)
```

```python
# extract the number of 'Situps'
situps = data['data'][:, 1]

#extract the value of 'Waist'
waist = data['target'][:, 1]
```

```python
# Create a DataFrame
df = pd.DataFrame({
    'Situps': situps,
    'Waist': waist
})

# Visualize the data
print(df)
```

```python
# Create and fit the model
model =  linear_model.LinearRegression()
X = df['Situps'].values.reshape(-1, 1)
y = df['Waist'].values
model.fit(X, y)

# Create a scatter plot
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Number of Situps')
plt.ylabel('Waistline (cm)')
plt.title('Situps vs. Waistline with Regression Line')
plt.show()
```
